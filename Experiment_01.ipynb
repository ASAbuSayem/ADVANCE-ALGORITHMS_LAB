# Experiment_01

**Aim: Write a program for Insertion sort and Selection Sort using count variable. Test the program, with three different Inputs (Reverse Order, Sorted and Unsorted arrays).**

**Algorithm: Insertion Sort with Count Variables
Steps:**

1. Start
2. Read number of elements n
3. Read array elements (space or comma separated)
4. Set outer = 0, inner = 0, comparison = 0
5. For i = 1 to n-1
   *   Increment outer
   *   Store arr[i] in key
   *   Set j = i - 1
6. While j >= 0 and arr[j] > key
   * Increment comparison
   * Shift element
   * Increment inner
   * Decrement j
7. Place key at correct position
8. Print sorted array and all counts
9. Stop

**Algorithm: Selection Sort with Count Variables
Steps:**

1. Start
2. Read array
3. Set outer = 0, inner = 0, comparison = 0
4. For i = 0 to n-2
   * Increment outer
   * Set min_index = i
5. For j = i+1 to n-1
   * Increment inner
   * Increment comparison
   * If arr[j] < arr[min_index], update min_index
6. Swap minimum element
7. Print sorted array and all counts
8. Stop

```python
# INPUT HANDLER
def read_array():
    data = input("Enter elements (space or comma separated): ")
    data = data.replace(",", " ")
    return list(map(int, data.split()))
```

```python
# INSERTION SORT
def insertion_sort(arr):
    outer = inner = comparisons = 0

    for i in range(1, len(arr)):
        outer += 1
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            comparisons += 1
            inner += 1
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return outer, inner, comparisons
```

```python
# SELECTION SORT
def selection_sort(arr):
    outer = inner = comparisons = 0
    n = len(arr)

    for i in range(n - 1):
        outer += 1
        min_index = i

        for j in range(i + 1, n):
            inner += 1
            comparisons += 1
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return outer, inner, comparisons
```

```python
# MAIN MENU
while True:
    print("\n===== Experiment–01 MENU =====")
    print("1. Insertion Sort")
    print("2. Selection Sort")
    print("3. Exit")

    algo = int(input("Select Algorithm: "))

    if algo == 3:
        print("Program Exited")
        break

    n = int(input("\nEnter number of elements: "))

    print("\nSelect Input Type:")
    print("1. Best Case (Already Sorted)")
    print("2. Worst Case (Reverse Order)")
    print("3. Average Case (Unsorted)")
    print("4. Exit")

    choice = int(input("Enter choice: "))

    if choice == 4:
        break

    arr = read_array()

    if choice == 1:
        arr = sorted(arr)
        case = "BEST CASE"
    elif choice == 2:
        arr = sorted(arr, reverse=True)
        case = "WORST CASE"
    elif choice == 3:
        case = "AVERAGE CASE"
    else:
        print("Invalid choice")
        continue

    print("\nOriginal Array:", arr)

    if algo == 1:
        outer, inner, comp = insertion_sort(arr)
        print("\nInsertion Sort Result:", arr)

    elif algo == 2:
        outer, inner, comp = selection_sort(arr)
        print("\nSelection Sort Result:", arr)

    print("\nAnalysis:")
    print("Case Type:", case)
    print("Outer Loop Count:", outer)
    print("Inner Loop Count:", inner)
    print("Total Comparisons:", comp)
```
