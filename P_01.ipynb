# 1.	Implement DFS and BFS procedures to search a node in a graph.

```python
from collections import deque
```

```python
# Graph creation
def create_graph():
    graph = {}
    n = int(input("Enter number of nodes: "))

    for i in range(n):
        node = input(f"Enter node {i+1} name: ")
        graph[node] = []

    e = int(input("Enter number of edges: "))
    print("Enter edges (source destination):")

    for i in range(e):
        u, v = input().split()
        graph[u].append(v)
        graph[v].append(u)   # because undirected graph

    return graph
```

```python
# DFS Search
def dfs(graph, start, target, visited):
    print(start, end=" ")  # show traversal
    visited.add(start)

    if start == target:
        return True

    for neighbor in graph[start]:
        if neighbor not in visited:
            if dfs(graph, neighbor, target, visited):
                return True

    return False
```

```python
# BFS Search
def bfs(graph, start, target):
    visited = set()
    queue = deque()

    queue.append(start)
    visited.add(start)

    while queue:
        node = queue.popleft()
        print(node, end=" ")

        if node == target:
            return True

        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return False
```

```python
# Dashboard
def main():
    graph = create_graph()

    while True:
        print("\n\n--- GRAPH SEARCH DASHBOARD ---")
        print("1. DFS Search")
        print("2. BFS Search")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            start = input("Enter starting node: ")
            target = input("Enter node to search: ")
            visited = set()
            print("\nDFS Traversal:")
            found = dfs(graph, start, target, visited)

            if found:
                print("\nNode Found using DFS")
            else:
                print("\nNode NOT Found using DFS")

        elif choice == '2':
            start = input("Enter starting node: ")
            target = input("Enter node to search: ")
            print("\nBFS Traversal:")
            found = bfs(graph, start, target)

            if found:
                print("\nNode Found using BFS")
            else:
                print("\nNode NOT Found using BFS")

        elif choice == '3':
            print("Exiting Program...")
            break

        else:
            print("Invalid choice! Try again.")


# Run the program
main()
```
