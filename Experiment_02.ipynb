**Aim: Write a program for Merge Sort and Quick Sort using count variable. Test the program, with three different Inputs (Reverse Order, Sorted and Unsorted arrays).**

**Algorithm: Merge Sort using Count Variable
Steps:**

1. Start
2. Read number of elements n
3. Read array elements
4. If array size > 1 then:

    a. Divide array into two halves

    b. Recursively apply Merge Sort on both halves

    c. Merge two sorted halves by comparing elements

    d. Increment comparison count during merge

5. Display sorted array
6. Display total comparison count
7. Stop

**Algorithm: Quick Sort using Count Variable
Steps:**

1. Start

2. Read array elements

3. Choose last element as pivot

4. Partition array such that:

   * elements smaller than pivot go left

   * elements greater than pivot go right

5. Increment comparison count during partition
6. Recursively apply Quick Sort on left and right subarrays
7. Display sorted array
8. Display total comparison count
9. Stop

```python
# MERGE SORT
def merge_sort(arr, count):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]

        count = merge_sort(L, count)
        count = merge_sort(R, count)

        i = j = k = 0

        while i < len(L) and j < len(R):
            count += 1
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

    return count
```

```python
# QUICK SORT
def quick_sort(arr, low, high, count):
    if low < high:
        pi, count = partition(arr, low, high, count)
        count = quick_sort(arr, low, pi - 1, count)
        count = quick_sort(arr, pi + 1, high, count)
    return count


def partition(arr, low, high, count):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        count += 1
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1, count
```

```python
n = int(input("Enter number of elements: "))
base_arr = list(map(int, input("Enter elements: ").split()))

while True:
    print("\nSelect Input Type:")
    print("1. Sorted Order")
    print("2. Reverse Order")
    print("3. Unsorted Order")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 4:
        print("Experiment Completed")
        break

    if choice == 1:
        arr = sorted(base_arr)
        case = "SORTED (BEST CASE)"
    elif choice == 2:
        arr = sorted(base_arr, reverse=True)
        case = "REVERSE (WORST CASE)"
    elif choice == 3:
        arr = base_arr.copy()
        case = "UNSORTED (AVERAGE CASE)"
    else:
        print("Invalid choice")
        continue

    print("\nInput Case:", case)
    print("Original Array:", arr)

    # Merge Sort
    ms_arr = arr.copy()
    ms_count = merge_sort(ms_arr, 0)

    # Quick Sort
    qs_arr = arr.copy()
    qs_count = quick_sort(qs_arr, 0, len(qs_arr) - 1, 0)

    print("\n--- MERGE SORT ---")
    print("Sorted Array:", ms_arr)
    print("Comparison Count:", ms_count)

    print("\n--- QUICK SORT ---")
    print("Sorted Array:", qs_arr)
    print("Comparison Count:", qs_count)
```
