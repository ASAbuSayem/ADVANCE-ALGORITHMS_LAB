**Lab-01:
 Insertion Sort using count variable.
Test the program with three different inputs
(Reverse order, Sorted order & Unsorted).**

```python
def insertion_sort(arr):
    count = 0

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            count += 1
            arr[j+1] = arr[j]
            j -= 1

        arr[j+1] = key

    return count
```

```python
# Main Program
n = int(input("Enter number of elements: "))

while True:
    print("\nSelect Input Type:")
    print("1. Best Case (Already Sorted)")
    print("2. Worst Case (Reverse Order)")
    print("3. Average Case (Unsorted)")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 4:
        print("Program Terminated.")
        break

    arr = []

    if choice == 1:
        print("Enter elements in Sorted Order:")
        for i in range(n):
            arr.append(int(input()))

    elif choice == 2:
        print("Enter elements in Reverse Order:")
        for i in range(n):
            arr.append(int(input()))

    elif choice == 3:
        print("Enter elements in Any Unsorted Order:")
        for i in range(n):
            arr.append(int(input()))

    else:
        print("Invalid Choice! Try Again.")
        continue

    print("\nOriginal Array:", arr)
    operations = insertion_sort(arr)
    print("Sorted Array:", arr)
    print("Total Operation Count:", operations)
```

**Lab-02: Selection Sort using count variable and analyze its complexity.**

```python
def selection_sort(arr):
    count = 0
    n = len(arr)

    for i in range(n):
        min_index = i

        for j in range(i+1, n):
            count += 1
            if arr[j] < arr[min_index]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

    return count
```

```python
# Main Program
n = int(input("Enter number of elements: "))

while True:
    print("\nSelect Input Type:")
    print("1. Best Case (Already Sorted)")
    print("2. Worst Case (Reverse Order)")
    print("3. Average Case (Unsorted)")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 4:
        print("Program Terminated.")
        break

    arr = []

    if choice == 1:
        print("Enter elements in Sorted Order:")
        for i in range(n):
            arr.append(int(input()))

    elif choice == 2:
        print("Enter elements in Reverse Order:")
        for i in range(n):
            arr.append(int(input()))

    elif choice == 3:
        print("Enter elements in Any Unsorted Order:")
        for i in range(n):
            arr.append(int(input()))

    else:
        print("Invalid Choice! Try Again.")
        continue

    print("\nOriginal Array:", arr)
    operations = selection_sort(arr)
    print("Sorted Array:", arr)
    print("Total Comparison Count:", operations)

    # Complexity Analysis
    max_ops = n * (n - 1) // 2
    print("Time Complexity ≈ O(n²)")
    print("Theoretical Max Operations:", max_ops)
```

**Lab-03: Implement Binary Search using count variable and analyze its time complexity.**

```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    count = 0

    while low <= high:
        count += 1
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, count
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, count
```

```python
# Main Program
n = int(input("Enter number of elements: "))

print("Enter elements in SORTED ORDER:")
arr = []
for i in range(n):
    arr.append(int(input()))

while True:
    print("\n--- Binary Search Menu ---")
    print("1. Search First Element")
    print("2. Search Middle Element")
    print("3. Search Last Element")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 4:
        print("Program Terminated.")
        break

    if choice == 1:
        key = arr[0]
        print("Searching First Element:", key)

    elif choice == 2:
        key = arr[n // 2]
        print("Searching Middle Element:", key)

    elif choice == 3:
        key = arr[n - 1]
        print("Searching Last Element:", key)

    else:
        print("Invalid Choice!")
        continue

    pos, operations = binary_search(arr, key)

    print("Element Found at Index:", pos)
    print("Total Comparisons:", operations)
    print("Time Complexity ≈ O(log n)")
```
