#**Aim: Implement the Prim's Algorithm to Find MST in a Graph**

**Algorithm: Primâ€™s Algorithm**

1. Start
2. Read vertex names
3. Read weighted edges
4. Select any starting vertex
5. Initialize:
   * Visited set
   * Minimum cost = 0
6. Repeatedly select the minimum weight edge:
   * One end visited
   * Other end unvisited
7. Add edge to MST
8. Mark new vertex as visited
9. Repeat until all vertices are visited
10. Display MST and total cost
11. Stop

```python
def prim(vertices, edges, start):
    visited = set([start])
    mst = []
    total_cost = 0

    while len(visited) < len(vertices):
        min_edge = None

        for u, v, w in edges:
            if (u in visited and v not in visited) or (v in visited and u not in visited):
                if min_edge is None or w < min_edge[2]:
                    min_edge = (u, v, w)

        if min_edge is None:
            break

        u, v, w = min_edge
        mst.append(min_edge)
        total_cost += w

        visited.add(v if u in visited else u)

    return mst, total_cost
```

```python
# Main Program
vertices = input("Enter vertex names (space separated, e.g. a b c d): ").split()
edges = []

E = int(input("Enter number of edges: "))
print("Enter edges in format: source destination weight")
print("Example: a b 2")

for _ in range(E):
    u, v, w = input().split()
    edges.append((u, v, int(w)))

start_vertex = input("Enter starting vertex: ")

while True:
    print("\n--- PRIM'S MST MENU ---")
    print("1. Find Minimum Spanning Tree")
    print("2. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        mst, cost = prim(vertices, edges, start_vertex)

        print("\nEdges in Minimum Spanning Tree:")
        for u, v, w in mst:
            print(f"{u} -> {v}  (Weight = {w})")

        print("Total Cost of MST:", cost)

    elif choice == 2:
        print("\nThank You!")
        print("Program executed successfully.")
        print("Developed by: Abu Sayem")
        break

    else:
        print("Invalid choice! Try again.")
```
