**Aim: Write a program to implement Binary Search using a count variable. Test the program and analyze the number of computations and time complexity.**

**Algorithm: Binary Search (with Count Variable)**

1. Start
2. Read number of elements n
3. Read array elements (must be sorted)
4. Read the element to be searched (key)
5. Set
   * low = 0
   * high = n - 1
   * count = 0

6. While low ≤ high, do:
   1. mid = (low + high) // 2
   2. Increment count by 1 (comparison happens)
   3. If arr[mid] == key
      * Element found
      * Print position and count
      * Stop

   4. Else if arr[mid] < key
      * Set low = mid + 1
   5. Else
      * Set high = mid - 1
7. If loop ends and element not found
    * Print “Element not found”
    * Print count
8. Stop

```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    count = 0

    while low <= high:
        mid = (low + high) // 2
        count += 1   # comparison count

        if arr[mid] == key:
            return mid, count
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, count
```

```python
# Main Program
n = int(input("Enter number of elements: "))
print("Enter elements in sorted order:")
arr = list(map(int, input().split()))

while True:
    print("\n--- Binary Search Menu ---")
    print("1. Search First Element")
    print("2. Search Middle Element")
    print("3. Search Last Element")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        key = arr[0]
        case = "First Element"

    elif choice == 2:
        key = arr[n // 2]
        case = "Middle Element"

    elif choice == 3:
        key = arr[-1]
        case = "Last Element"

    elif choice == 4:
        print("Exiting Program...Bye")
        break

    else:
        print("Invalid Choice!")
        continue

    pos, operations = binary_search(arr, key)

    print(f"\nSearching {case}: {key}")
    print(f"Element Found at Index: {pos}")
    print(f"Total Comparisons: {operations}")
    print("Time Complexity ≈ O(log n)")
```
